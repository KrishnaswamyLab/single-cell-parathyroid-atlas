```python
import scanpy, scprep, graphtools, magic, phate
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import cluster, preprocessing, decomposition
import pandas as pd
from sklearn.preprocessing import scale
```

```python
anndata_primate_raw = scanpy.read_h5ad('../data/processed_files/4_primate_parathyroid_cells_all_genes.h5ad')

# set main data array to be the batch mean corrected counts to avoid visualizing differences between batches
anndata_primate = anndata_primate_raw.copy()
anndata_primate.X = anndata_primate.layers['X_batch_mean']
```

```python
# run MAGIC with t=1 as a way to get batch-corrected counts based on corrected underlying graph

primate_graph = graphtools.Graph(anndata_primate.obsm["X_pca"], use_pygsp=True, n_pca=None,
                                 random_state=42, sample_idx=anndata_primate.obs.batch.astype(int), bandwidth_scale=0.5)

magic_op = magic.MAGIC(random_state=42, t=1)
magic_op.graph = primate_graph
anndata_primate.layers['X_magic'] = magic_op.transform(anndata_primate.to_df().values) 
anndata_primate.X = anndata_primate.layers['X_magic']
```

```python
sns.set_palette('Set2')
```

# Clustering gene trajectories (all replicates)

```python
genes = np.zeros((anndata_primate.to_df().shape[1], 101))
for i, time in enumerate(np.arange(0,1.01,0.01)):
    genes[:, i] = anndata_primate.to_df()[np.round(anndata_primate.obs['pseudotime'], 2) == np.round(time, 2)].mean(axis=0)
```

### genes by time
### scale(time by genes)

```python
all_values_scaled = pd.DataFrame(preprocessing.scale(np.array(genes).T).T, index=anndata_primate.to_df().columns)
all_values_scaled_rolling = pd.DataFrame(all_values_scaled).rolling(window=20, center=True, axis=1).mean().T.dropna().T
```

```python
markers = ['PTH', 'CHGA', 'GCM2', 'COX1', 'COX2', 'COX4I1']
```

```python
fig = plt.figure(figsize=(7,4))
for marker in markers:
    sns.lineplot(x=np.array(list(range(all_values_scaled_rolling.shape[1]))) / 100,
                 y=all_values_scaled_rolling.loc[marker], label=marker, linewidth=3)
plt.legend(bbox_to_anchor=(1.05, 1), fontsize=14)
plt.ylabel('Scaled Expression', fontsize=14)
plt.xlabel('Pseudotime', fontsize=14)
plt.tight_layout()
fig.savefig('figures/primate_parathyroid/pseudotime_marker_expression.png', dpi=200)
```

## Gene clusters

```python
km = cluster.KMeans(random_state=42, n_clusters=6)
km.fit(all_values_scaled_rolling)
```

```python
# rename clusters based on ordering of peak in time

conversions = {5:0, 2:1, 3:2, 4:3, 1:4, 0:5}
anndata_primate.var['cluster'] = [conversions[x] + 1 for x in km.labels_]
```

```python
def plot_clusters(genes, cl, values_scaled_rolling):
    ulabels = np.unique(cl)
    
    df = values_scaled_rolling.T.reset_index().melt(id_vars=['index'], value_vars=[*genes])
    
    df['cluster'] = np.repeat(cl, values_scaled_rolling.shape[1])
    df.columns = ['time', 'variable', 'scaled expression', 'cluster']
    
    g = sns.relplot(x='time', 
                    y='scaled expression', 
                    hue='cluster', 
                    kind='line',
                    col='cluster',
                    data=df, 
                    palette='tab10', 
                    legend='full', 
                    col_wrap=3,
                    errorbar='sd',)

    plt.savefig('figures/primate_parathyroid/gene_clusters.png', dpi=600)
```

```python
plot_clusters(anndata_primate.to_df().columns, np.array(anndata_primate.var['cluster']), all_values_scaled_rolling)
```

```python
anndata_primate.var.to_csv('../data/primate_pseudotime_gene_clusters.csv')
```

## Plotting genes

```python
markers = ['PTH', 'CHGA', 'GCM2', 'COX1', 'COX2', 'COX4I1', 'RARRES2']
```

```python
ulabels = np.unique(anndata_primate.var.cluster - 1)

lut = {i: plt.get_cmap("tab10")(i) for i in range(6)}
row_colors = (
    pd.Series(anndata_primate.var.cluster - 1, index=anndata_primate.var.index).map(lut).rename("Cluster Order")
)
clx = sns.clustermap(all_values_scaled_rolling, col_cluster=False, row_colors=row_colors, cmap="inferno", robust=True)
ax = clx.ax_heatmap
```

```python
np.save('../data/gene_dynamics_ordering.npy', np.array(clx.dendrogram_row.reordered_ind))
```

```python
def plot_heatmap(genes, cl, df, to_display):
    
    cluster_order = [0, 1, 2, 3, 4, 5]
    reorder2 = []
    reorder = np.array(clx.dendrogram_row.reordered_ind)
    for c in cluster_order:
        reorder2.extend(reorder[c == cl[reorder]])
    reorder2 = np.array(reorder2)
    ss = df.T.iloc[reorder2]
    rc = row_colors[reorder2]
    clxx = sns.clustermap(
        ss, col_cluster=False, row_cluster=False, cmap="inferno", robust=True, row_colors=rc
    )
    
    ax = clxx.ax_heatmap
    names = anndata_primate.var.index
    reordered_names =names[reorder2]
    ax.set_yticks(np.arange(len(names))[reordered_names.isin(to_display)])
    ticks = ax.set_yticklabels(reordered_names[reordered_names.isin(to_display)])
    clxx.ax_heatmap.set_xticklabels([])
    clxx.ax_heatmap.set_xticks([])
    
    clxx.savefig('figures/primate_parathyroid/pseudotime_all_heatmap.png', dpi=200)
    
    
plot_heatmap(anndata_primate.to_df().columns, anndata_primate.var.cluster - 1, all_values_scaled_rolling.T,
             anndata_primate[:, markers].var.cluster.sort_values().index)
```
