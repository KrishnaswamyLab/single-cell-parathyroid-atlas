```python
import numpy as np
import pandas as pd
import phate, scprep
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import seaborn as sns
import scipy
```

## Visualization of mapping

```python
with open('/home/av622/parathyroid_v2/processed_files/5_primate_parathyroid_cells_all_genes_batch_corrected.npz', 'rb') as f:
    npzfile = np.load(f, allow_pickle=True)
    primate_magan_data = npzfile['data']
    primate_genes = npzfile['genes']
    
primate_magan_data = pd.DataFrame(primate_magan_data, columns=primate_genes)
primate_magan_data = primate_magan_data.iloc[:, ~primate_magan_data.columns.duplicated()]
```

```python
with open('/home/av622/parathyroid_v2/processed_files/5_scMMGAN_human_to_primate.npz', 'rb') as f:
    npzfile = np.load(f, allow_pickle=True)
    human_magan_data = npzfile['human_to_primate']
    cells = npzfile['cells']
    genes = npzfile['genes']
    human_samples = npzfile['samples']
    
human_magan_data = pd.DataFrame(human_magan_data, index=cells, columns=genes)
human_magan_data = human_magan_data.iloc[:, ~human_magan_data.columns.duplicated()]
```

```python
phate_op = phate.PHATE(random_state=42, use_pygsp=True, knn=10, verbose=True, t=20)
primate_phate = phate_op.fit_transform(primate_magan_data)
primate_phate[:, 0] = -1* primate_phate[:, 0]
```

```python
human_phate = phate_op.transform(human_magan_data)
human_phate[:, 0] = -1* human_phate[:, 0]
```

```python
fig, ax = plt.subplots(1,1, figsize=(4.5,4))
scprep.plot.scatter2d(primate_phate, c='lightgray', ticks=True, ax=ax, title='Human parathyroid cells\nmapped to primate cells'), 
scprep.plot.scatter2d(human_phate, c=human_samples, cmap='Spectral', ticks=None, label_prefix="PHATE", ax=ax,
                     legend_loc=(1.05,0),)
plt.tight_layout()
fig.savefig(f'figures/human_parathyroid/human_samples_mapped.png', dpi=200)
```

```python
for sample in ['Y11', 'Y13', 'Y7', 'Y9']:
    density = gaussian_kde(human_phate[human_samples == sample].T)(human_phate[human_samples == sample].T)
    fig, ax = plt.subplots(1,1, figsize=(5,5))
    scprep.plot.scatter2d(primate_phate, c='lightgray',
                          ticks=None, ax=ax, title=f'{sample} density')
    scprep.plot.scatter2d(human_phate[human_samples == sample], c=density, label_prefix='PHATE',
                          ticks=None, ax=ax, cmap='inferno')

    fig.savefig(f'figures/human_parathyroid/human_{sample}_mapped.png', dpi=200)
```

## Correlations between input and generated human

```python
anndata_human = scanpy.read_h5ad('../data/processed_files/4_human_parathyroid_cells_all_genes.h5ad')
```

```python
sns.set(style='white', font_scale=1.4)
```

```python
intersection_genes = set(anndata_human.var['PrimateEnsemblID'].tolist()).intersection(human_magan_data.columns)
```

```python
df = pd.DataFrame(anndata_human.layers['X_magic'].todense(), columns=anndata_human.var['PrimateEnsemblID'])
correlation = {}
for gene in intersection_genes:
    correlation[gene] = scipy.stats.spearmanr(df[gene], human_magan_data[gene]).correlation
correlation = pd.DataFrame((correlation.keys(), correlation.values())).T
correlation.columns = ['Gene', 'Correlation']
correlation = correlation.set_index('Gene')

df = pd.concat((correlation.sort_values('Correlation')[:5], correlation.sort_values('Correlation', ascending=False)[:5]))
genes = ['PTH', 'CHGA', 'RARRES2', 'GCM2', 'KL', 'MT-CO1', 'MT-CO2', 'VDR', 'PIN1']
df = pd.concat((df, correlation.loc[genes])).sort_values('Correlation')
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
sns.barplot(y=df.index, orient='h', x=df['Correlation'], palette=['lightgray']*5 + ['#9e0142']*len(genes) + ['lightgray']*5)
plt.ylabel('Correlation(human, scMMGAN-generated human)')
plt.tight_layout()
fig.savefig('figures/human_parathyroid/top_pos_and_neg_correlation.png', dpi=600)
```
