```python
import numpy as np
import pandas as pd
import scanpy
from sklearn import metrics, preprocessing
import seaborn as sns
import phate
import matplotlib.pyplot as plt
import scprep
```

```python
with open('../data/processed_files/5_primate_parathyroid_cells_all_genes_batch_corrected.npz', 'rb') as f:
    npzfile = np.load(f, allow_pickle=True)
    primate_magan_data = npzfile['data']
    primate_genes = npzfile['genes']
    
primate_magan_data = pd.DataFrame(primate_magan_data, columns=primate_genes)
primate_magan_data = primate_magan_data.iloc[:, ~primate_magan_data.columns.duplicated()]
```

```python
df_human = scanpy.read_h5ad('../data/processed_files/5_human_parathyroid_cells_all_genes_batch_corrected.h5ad')
```

```python
with open('../data/processed_files/6_scMMGAN_human_to_primate.npz', 'rb') as f:
    npzfile = np.load(f)
    human_magan_data = npzfile['human_to_primate']
    
human_magan_data = pd.DataFrame(human_magan_data, index=df_human.obs_names, columns=primate_genes)
human_magan_data = human_magan_data.iloc[:, ~human_magan_data.columns.duplicated()]
```

```python
data_ref = pd.DataFrame(df_human.to_df().values, columns=df_human.var['PrimateEnsemblID'].tolist(), index=df_human.obs_names)
data_ref = data_ref.loc[:, data_ref.sum(axis=0) != 0]
genes = np.array(list(primate_magan_data.columns))

intersection_genes = np.array(list(set(genes).intersection(data_ref.columns)))
data_ref[intersection_genes] = (data_ref[intersection_genes] - data_ref[intersection_genes].mean(axis=0) + primate_magan_data[intersection_genes].mean(axis=0))
```

```python
anndata_primate_raw = scanpy.read_h5ad('../data/processed_files/4_primate_parathyroid_cells_all_genes.h5ad')
pseudotime = anndata_primate_raw.obs['pseudotime']
```

```python
from scipy.spatial.distance import cdist
dist = cdist(human_magan_data, primate_magan_data)
nearest = np.argsort(dist, axis=1)[:, 0]
df_human.obs['magan_mapped_time'] = pseudotime[nearest].values
```

## Visualize results of encoded NHP, output human as gene expression heatmaps (ortholog-mapped)

```python
phate_op = phate.PHATE(random_state=42, use_pygsp=True, knn=10, verbose=True, t=20)
primate_phate = phate_op.fit_transform(primate_magan_data)
human_phate = phate_op.transform(human_magan_data)
```

```python
primate_phate = primate_phate * [-1,1]
human_phate = human_phate * [-1,1]
```

```python
def my_ceil(a, precision=0):
    return np.round(a + 0.5 * 10**(-precision), precision)
```

```python
primate_tp = my_ceil(pseudotime, 1)
```

```python
out = pd.concat((primate_magan_data[intersection_genes], data_ref[intersection_genes], human_magan_data[intersection_genes]))
out = preprocessing.scale(out)
encoded_nhp = out[:primate_magan_data.shape[0]]
input_human = out[primate_magan_data.shape[0]:primate_magan_data.shape[0]+data_ref.shape[0]]
output_human = out[primate_magan_data.shape[0]+data_ref.shape[0]:]
```

```python
df = pd.DataFrame(columns = intersection_genes)
```

```python
for timepoint in np.linspace(0.1, 1.0, 10):
    timepoint = np.round(timepoint, 1)
    df.loc[f'NHP {timepoint}'] = np.mean(encoded_nhp[primate_tp == timepoint], axis=0)
    df.loc[f'Human {timepoint}'] = np.mean(output_human[df_human.obs['magan_mapped_time_bins'] == timepoint], axis=0)
    for sample in ['Y7', 'Y9', 'Y11', 'Y13']:
        df.loc[f'{sample} {timepoint}'] = np.mean(output_human[(df_human.obs['magan_mapped_time_bins'] == timepoint) & (df_human.obs['sample'] == sample)], axis=0)

df = df.astype(float)
```

```python
g= sns.clustermap(df, row_cluster=False, col_cluster=True, figsize=(10,20), cmap='RdBu_r', xticklabels=False)
ax = g.ax_heatmap
ax.set_xlabel("All genes with mapped orthologs")
ax.set_ylabel("Samples pseudobulked over time")
g.savefig('figures/human_parathyroid/genome_scale_heatmap_mapping.png')
```

## Sample-specific gene trends

```python
sample = 'Y7'
```

```python
human_magan_sample = human_magan_data[df_human.obs['sample'] == sample]
```

```python
genes = np.zeros((human_magan_sample.shape[1], 101))
for i, time in enumerate(np.arange(0,1.01,0.01)):
    genes[:, i] = human_magan_sample[np.round(df_human.obs['magan_mapped_time'][df_human.obs['sample'] == sample], 2) == np.round(time, 2)].mean(axis=0)
```

```python
all_values_scaled = pd.DataFrame(preprocessing.scale(np.array(genes).T).T, index=human_magan_sample.columns)
all_values_scaled_rolling = pd.DataFrame(all_values_scaled).rolling(window=2, center=True, axis=1).mean().T.dropna().T
```

## Gene clusters

```python
gene_clusters = pd.read_csv('../data/primate_pseudotime_gene_clusters.csv', index_col=0)
gene_clusters.index = anndata_primate_raw.var['reannotated_gene_names'].values
gene_clusters = gene_clusters[~gene_clusters.index.duplicated()].loc[human_magan_data.columns]['cluster']
```

```python
markers = ['PTH', 'CHGA', 'GCM2', 'MT-CO1', 'MT-CO2', 'RARRES2', 'COX4I1']
```

```python
ulabels = np.unique(gene_clusters - 1)

lut = {i: plt.get_cmap("tab10")(i) for i in range(6)}
row_colors = (
    pd.Series(gene_clusters - 1, index=human_magan_data.columns).map(lut).rename("Cluster Order")
)
```

```python
dendrogram_order = np.load('../data/gene_dynamics_ordering.npy')
order = anndata_primate_raw.var.iloc[dendrogram_order]['reannotated_gene_names']
order = order[~order.duplicated()].values
```

```python
_, order_id = np.where(order[:,None] == human_magan_data.columns.values)
```

```python
def plot_heatmap(genes, cl, df, to_display):
    
    cluster_order = [0, 1, 2, 3, 4, 5]
    reorder2 = []
    reorder = np.array(order_id)
    for c in cluster_order:
        reorder2.extend(reorder[c == cl[reorder]])
    reorder2 = np.array(reorder2)
    ss = df.T.iloc[reorder2]
    rc = row_colors[reorder2]
    clxx = sns.clustermap(
        ss, col_cluster=False, row_cluster=False, cmap="inferno", robust=True, row_colors=rc
    )
    
    ax = clxx.ax_heatmap
    names = genes
    reordered_names =names[reorder2]
    ax.set_yticks(np.arange(len(names))[reordered_names.isin(to_display)])
    ticks = ax.set_yticklabels(reordered_names[reordered_names.isin(to_display)])
    clxx.ax_heatmap.set_xticklabels([])
    clxx.ax_heatmap.set_xticks([])
    
    clxx.savefig(f'figures/human_parathyroid/pseudotime_all_heatmap_{sample}.png', dpi=200)
    
    
plot_heatmap(human_magan_data.columns, gene_clusters.values - 1, all_values_scaled_rolling.T,
             gene_clusters.loc[markers].sort_values().index)
```
