```python
import scanpy as sc
import scprep
import matplotlib.pyplot as plt
import numpy as np
```

```python
sdata = sc.read_h5ad('../data/processed_files/4_human_parathyroid_cells_all_genes.h5ad')
```

```python
scprep.plot.scatter2d(sdata.obsm["X_phate_mnn"], c=sdata.obs['sample'],
                          legend_loc=(1.05,0),
                          ticks=None, title='Human parathyroid cells\nSamples',
                          label_prefix='PHATE', dpi=200, figsize=(4.8,4),
                          filename=f'figures/human_parathyroid/samples.png')
```

```python
for marker in ['CASR', 'PTH', 'CHGA', 'GCM2', 'COX4I1', 'MT-CO2', 'RARRES2']:
    scprep.plot.scatter2d(sdata.obsm["X_phate_mnn"], c=sdata.to_df()[marker],
                          legend_loc=(1.05,0),
                          ticks=None, title=marker,
                          label_prefix='PHATE', dpi=200, figsize=(4.4,4),
                          filename=f'figures/human_parathyroid/{marker}.png')
```

## Lowly-expressed genes

```python
genes= ['FGFR2','LRP6','MTOR','PAX1','PAX9','TBX1','GCM2','SIX1','SIX4','EYA1','GATA3', 'FZD5']

fig, ax = plt.subplots(4,3, figsize=(12,16)); ax=ax.flatten()

for i,gene in enumerate(genes):
    keepCells = scprep.filter.filter_values(sdata, values=(sdata.to_df()[gene]), percentile=90, keep_cells='above')
    perc = np.percentile(keepCells.to_df()[gene], 90)
    
    scprep.plot.scatter2d(sdata.obsm['X_phate_mnn'], c='lightgray',
                          ticks=None, ax=ax[i], title=gene, s=10)
    scprep.plot.scatter2d(keepCells.obsm['X_phate_mnn'], c=keepCells.to_df()[gene], ticks=None, ax=ax[i], title=gene, s=20,
                         label_prefix='PHATE', vmax=perc) 
    
plt.tight_layout()
fig.savefig('figures/human_parathyroid/top_10%_maturation.png', dpi=200)
```
