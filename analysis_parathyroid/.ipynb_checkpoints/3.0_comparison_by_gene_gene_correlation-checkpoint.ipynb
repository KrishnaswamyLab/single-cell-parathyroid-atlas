```python
import scanpy as sc
import scprep
import phate
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphtools
from scipy import stats, spatial
import magic
from sklearn import preprocessing
import leidenalg
```

## Get denoised counts per sample

```python
sdata = sc.read_h5ad('../data/processed_files/4_primate_parathyroid_cells_all_genes.h5ad')
samples = sdata.obs['sample'].cat.categories

for sample in samples:
    sdata_sample = sdata[sdata.obs['sample'] == sample]
    phate_op = phate.PHATE(random_state=42, verbose=False)
    data_phate = phate_op.fit_transform(sdata_sample.X)
    sdata_sample.obsm["X_pca"] = phate_op.graph.data_nu
    
    magic_op = magic.MAGIC(random_state=0)
    magic_op.graph = phate_op.graph
    sdata_sample.layers["X_magic"] = magic_op.transform(sdata_sample.to_df())
    
    sdata_sample.write(f'../data/processed_files/6_primate_{sample}_parathyroid_cells_all_genes.h5ad')
```

```python
sdata = sc.read_h5ad('../data/processed_files/4_human_parathyroid_cells_all_genes.h5ad')
samples = sdata.obs['sample'].cat.categories

for sample in samples:
    sdata_sample = sdata[sdata.obs['sample'] == sample]
    phate_op = phate.PHATE(random_state=42, verbose=False)
    data_phate = phate_op.fit_transform(sdata_sample.X)
    sdata_sample.obsm["X_pca"] = phate_op.graph.data_nu
    
    magic_op = magic.MAGIC(random_state=0)
    magic_op.graph = phate_op.graph
    sdata_sample.layers["X_magic"] = magic_op.transform(sdata_sample.to_df())
    
    sdata_sample.write(f'../data/processed_files/6_human_{sample}_parathyroid_cells_all_genes.h5ad')
```

## Correlations within samples

```python
sample = 'primate_sample1'
anndata_sample = sc.read_h5ad(f'../data/processed_files/per_sample/6_{sample}_parathyroid_cells_all_genes.h5ad')
reannotation = pd.read_csv('../data/reannotated_gene_names.csv', index_col=0)
mapping = dict(zip(reannotation['X'], reannotation['reannotated_gene_names']))
anndata_sample.var['reannotation'] = [mapping[x] for x in anndata_sample.var_names]

primate_gene_names = anndata_sample.var['reannotation']
```

```python
sample = 'human_Y7'
anndata_sample = sc.read_h5ad(f'../data/processed_files/per_sample/6_{sample}_parathyroid_cells_all_genes.h5ad')        
    
human_gene_names = sc.read_h5ad(f'../data/processed_files/4_human_parathyroid_cells_all_genes.h5ad').var['PrimateEnsemblID']
human_gene_names = human_gene_names.loc[anndata_sample.var_names]
anndata_sample.var['PrimateEnsemblID'] = human_gene_names 

human_gene_names = anndata_sample.var['PrimateEnsemblID']
```

```python
intersection_genes = np.array(list(set(primate_gene_names).intersection(human_gene_names)))
```

```python
human_samples = ['human_Y7', 'human_Y9', 'human_Y11', 'human_Y13']
primate_samples = ['primate_sample1', 'primate_sample2', 'primate_sample3', 'primate_sample4']
```

```python
primate_corr = {}
```

```python
for sample in primate_samples:
    print(sample)
    anndata_sample = sc.read_h5ad(f'../data/processed_files/per_sample/6_{sample}_parathyroid_cells_all_genes.h5ad')

    ## get clusters
    G = graphtools.Graph(anndata_sample.obsm["X_pca"], use_pygsp=True, n_pca=None, random_state=42)
    partition = leidenalg.find_partition(G.to_igraph(), leidenalg.RBConfigurationVertexPartition, seed=0,
                                         weights="weight", resolution_parameter=10, n_iterations=-1)
    anndata_sample.obs[f"clusters_res_1"] = np.array(partition.membership).astype('str')

    ## get pseudobulk with magic'd mean
    pbs = []
    for cluster in anndata_sample.obs['clusters_res_1'].unique():
        anndata_sample_subset = anndata_sample[anndata_sample.obs['clusters_res_1'] == cluster]
        
        pseudobulk_anndata = sc.AnnData(X = anndata_sample_subset.layers['X_magic'].mean(axis=0).reshape(1,-1), var=anndata_sample_subset.var[[]])
        pseudobulk_anndata.obs['sample'] = sample
        pseudobulk_anndata.obs['cluster'] = int(cluster)
        pbs.append(pseudobulk_anndata)
    
    pb = sc.concat(pbs)
    pb.obs_names = anndata_sample.obs['clusters_res_1'].unique()
    
    pb.var_names = primate_gene_names.values
    pb = pb.to_df().loc[:, ~pb.to_df().columns.duplicated()][intersection_genes]
    primate_corr[sample] = pb.corr()
```

```python
human_corr = {}
for sample in human_samples:
    print(sample)
    anndata_sample = sc.read_h5ad(f'../data/processed_files/per_sample/6_{sample}_parathyroid_cells_all_genes.h5ad')

    ## get clusters
    G = graphtools.Graph(anndata_sample.obsm["X_pca"], use_pygsp=True, n_pca=None, random_state=42)
    partition = leidenalg.find_partition(G.to_igraph(), leidenalg.RBConfigurationVertexPartition, seed=0,
                                         weights="weight", resolution_parameter=10, n_iterations=-1)
    anndata_sample.obs[f"clusters_res_1"] = np.array(partition.membership).astype('str')
    print (anndata_sample.obs[f"clusters_res_1"].unique())

    ## get pseudobulk with magic'd mean
    pbs = []
    for cluster in anndata_sample.obs['clusters_res_1'].unique():
        anndata_sample_subset = anndata_sample[anndata_sample.obs['clusters_res_1'] == cluster]
        
        pseudobulk_anndata = sc.AnnData(X = anndata_sample_subset.layers['X_magic'].mean(axis=0).reshape(1,-1), var=anndata_sample_subset.var[[]])
        pseudobulk_anndata.obs['sample'] = sample
        pseudobulk_anndata.obs['cluster'] = int(cluster)
        pbs.append(pseudobulk_anndata)
    
    pb = sc.concat(pbs)
    pb.obs_names = anndata_sample.obs['clusters_res_1'].unique()
    
    pb.var_names = human_gene_names.values
    pb = pb.to_df().loc[:, ~pb.to_df().columns.duplicated()][intersection_genes]
    human_corr[sample] = pb.corr()
```

```python
corr = {**human_corr, **primate_corr}
```

```python
corr_of_corr = np.zeros((8,8))

for i,(sample1, correlation1) in enumerate(corr.items()):
    for j, (sample2, correlation2) in enumerate(corr.items()):
        if i ==j:
            continue
        if corr_of_corr[i][j] != 0:
            continue
            
        corr_of_corr[i][j] = pd.concat((pd.Series(correlation1.values[np.triu_indices(len(intersection_genes))]), pd.Series(correlation2.values[np.triu_indices(len(intersection_genes))])), axis=1).corr().loc[0][1]
        corr_of_corr[j][i] = corr_of_corr[i][j]
        
        print(sample1, sample2, corr_of_corr[i][j])
        print(corr_of_corr)
```

```python
for i in range(8):
    corr_of_corr[i][i] = 1
```

```python
sns.set(font_scale=2)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr_of_corr, annot=True)
ax.set_xticklabels(corr.keys())
ax.set_yticklabels(corr.keys())

plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.tight_layout()
fig.savefig('figures/correlation_of_correlation.png', dpi=200)
```
