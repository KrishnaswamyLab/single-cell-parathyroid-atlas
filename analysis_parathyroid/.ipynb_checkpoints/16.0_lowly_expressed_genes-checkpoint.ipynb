```python
import scanpy, scprep, graphtools, magic, phate
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import cluster, preprocessing, decomposition
import pandas as pd
from localization import *
from collections import defaultdict, Counter
from sklearn.preprocessing import scale
```

```python
anndata_primate = scanpy.read_h5ad('processed_files/4_human_parathyroid_cells_all_genes.h5ad')
anndata_primate.X = anndata_primate.raw.X
```

```python
genes= ['FGFR2','LRP6','MTOR','PAX1','PAX9','TBX1','GCM2','SIX1','SIX4','EYA1','GATA3', 'FZD5']

fig, ax = plt.subplots(4,3, figsize=(12,16)); ax=ax.flatten()

for i,gene in enumerate(genes):
    keepCells = scprep.filter.filter_values(anndata_primate, values=(anndata_primate.to_df()[gene]), percentile=90, keep_cells='above')
    perc = np.percentile(keepCells.to_df()[gene], 90)
    
    scprep.plot.scatter2d(anndata_primate.obsm['X_phate_mnn'], c='lightgray',
                          ticks=None, ax=ax[i], title=gene, s=10)
    scprep.plot.scatter2d(keepCells.obsm['X_phate_mnn'], c=keepCells.to_df()[gene], ticks=None, ax=ax[i], title=gene, s=20,
                         label_prefix='PHATE', vmax=perc) 
    
plt.tight_layout()
fig.savefig('paper_figures/human_parathyroid/top_10%_maturation.png', dpi=200)
```

```python
heatmap = pd.DataFrame(index=genes, columns=np.arange(0,1.01,0.01))
```

```python
for i,gene in enumerate(genes):
    cell_mask = scprep.filter.filter_values(list(range(anndata_primate.n_obs)), values=(anndata_primate.to_df()[gene]),
                                                  percentile=90, keep_cells='above')
    
    for i, time in enumerate(np.arange(0,1.01,0.01)):
        time_mask = np.where(np.round(anndata_primate.obs['pseudotime'], 1) == np.round(time, 1))[0]
        ids = list(set(time_mask).intersection(cell_mask))
        heatmap.loc[gene][time] = anndata_primate[ids, gene].X.mean()
```

```python
all_values_scaled = pd.DataFrame(preprocessing.scale(np.array(heatmap).T).T, index=genes)
all_values_scaled_rolling = all_values_scaled.rolling(window=20, center=True, axis=1).mean().T.dropna().T
```

```python
fig = plt.figure(figsize=(7,4))
for gene in genes:
    sns.lineplot(x=np.array(list(range(all_values_scaled_rolling.shape[1]))) / 100,
                 y=all_values_scaled_rolling.loc[gene], label=gene, linewidth=3)
plt.legend(bbox_to_anchor=(1.05, 1), fontsize=14)
plt.ylabel('Scaled Expression', fontsize=14)
plt.xlabel('Pseudotime', fontsize=14)
plt.tight_layout()
```

```python
def plot_heatmap(df, adata, genes, savepath=None, figsize=(10, 7)):
    clx = sns.clustermap(
        df.T,
        col_cluster=False,
        row_cluster=False,
        cmap="inferno",
        robust=True,
        # dendrogram_ratio=(0.2, 0.01),
        figsize=figsize,
    )
    
    ax = clx.ax_heatmap
    clx.ax_heatmap.set_yticklabels(
        np.array(genes), rotation=0, 
    )
    clx.ax_heatmap.set_xticklabels([])
    clx.ax_heatmap.set_xticks([])
    clx.ax_row_dendrogram.set_visible(False)
    if savepath is not None:
        clx.savefig(savepath, dpi=600)
    return clx
```

```python
clx = plot_heatmap(all_values_scaled_rolling.T[genes], anndata_primate, genes,
                   figsize=(5, 4.3), savepath='paper_figures/human_parathyroid/markers_heatmap.png')
```
