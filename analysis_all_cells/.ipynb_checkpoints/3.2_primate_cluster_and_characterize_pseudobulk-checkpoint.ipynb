```python
import scprep, scipy, graphtools, phate
import scanpy as sc
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import scale
from scipy import io
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
import matplotlib.pyplot as plt
```

```python
sdata = sc.read_h5ad('intermediate_files/3_primate_all_cells_clustering.h5ad')
```

```python
cluster_conv = {'0': '2', '1': '1', '2':'3', '3':'4', '4':'5'}
sdata.obs['clusters_res_0.1'] = [cluster_conv[x] for x in sdata.obs['clusters_res_0.1']]
```

```python
sdata.layers['decontx'] = io.mmread('intermediate_files/2_decontx_primate_data.mtx').T
sdata.layers['decontx'] = scipy.sparse.csr_matrix(sdata.layers['decontx'])
sdata.layers['decontx'] = sdata.layers['decontx'].astype(int)
```

```python
res = 0.1
scprep.plot.scatter2d(sdata.obsm['X_phate_mnn_bmc'], c=sdata.obs[f'clusters_res_{res}'], legend_loc=(1.05,0), ticks=None,
                      label_prefix='PHATE', dpi=200, figsize=(5,4),  title='Primate all cells\nClusters')
```

```python
cluster_conv = {'1': '1', '2': '1', '3':'2', '4':'3', '5':'4'}
sdata.obs[f'clusters_combined'] = [cluster_conv[x] for x in sdata.obs['clusters_res_0.1']]
```

```python
res = 0.1
scprep.plot.scatter2d(sdata.obsm['X_phate_mnn_bmc'], c=sdata.obs[f'clusters_combined'], legend_loc=(1.05,0), ticks=None,
                      label_prefix='PHATE', dpi=200, figsize=(5,4),  title='Primate all cells\nClusters', filename='revision_results//primate_all_cells/clusters_combined.png')
```

```python
sdata_human_annot = sdata[:, ~sdata.var['primate_annotated']]
#sdata_human_annot.var_names = sdata_human_annot.var['reannotated_gene_names']
```

```python
sdata_human_annot.var_names = sdata_human_annot.var['gene_ids']
```

```python
sdata = sdata_human_annot
```

```python
pbs = []
for sample in sdata.obs['sample'].unique():
    for cluster in sdata.obs['clusters_combined'].unique():
        sdata_subset = sdata[sdata.obs['sample'] == sample]
        samp_cell_subset = sdata_subset[sdata_subset.obs['clusters_combined'] == cluster]
        samp_cell_subset.X = samp_cell_subset.layers['decontx']

        rep_adata = sc.AnnData(X = samp_cell_subset.X.sum(axis=0), var=samp_cell_subset.var[[]])
        rep_adata.obs['sample'] = sample
        rep_adata.obs['cluster'] = int(cluster)
        pbs.append(rep_adata)
```

```python
pb = sc.concat(pbs)
pb.obs_names = range(16)
```

```python
counts = pd.DataFrame(pb.X, columns = pb.var_names)
```

```python
for cluster in [1,2,3,4]:
    pb.obs[f'in_cluster_{cluster}_vs_1'] = pb.obs['cluster'].isin([cluster])
```

```python
for sample in ['sample2', 'sample3', 'sample4']:
    pb.obs[f'in_{sample}_vs_sample1'] = pb.obs['sample'].isin([sample])
```

```python
for cluster in [1,2,3,4]:
    dds = DeseqDataSet(counts=counts, metadata=pb.obs, 
                   design_factors=[f'in_cluster_{cluster}_vs_1', 'in_sample2_vs_sample1', 'in_sample3_vs_sample1', 'in_sample4_vs_sample1'])
    dds.deseq2()
    stat_res = DeseqStats(dds, n_cpus=8, contrast=(f'in-cluster-{cluster}-vs-1', 'True', 'False'))
    stat_res.summary()
    df = stat_res.results_df
    df = df[(df['log2FoldChange'] > 0) & (df['padj'] < 0.05)].sort_values('log2FoldChange', ascending=False)

    df['gene_name'] = sdata_human_annot.var.loc[df.index]['reannotated_gene_names']
    df.to_csv(f'revision_results/primate_all_cells/pseudobulk_cluster_{cluster}_vs_rest_combine_PTH_clusters.csv')
```

## Visualization

```python
clusters = 'clusters_res_0.1'; suffix = ''
clusters='clusters_combined'; suffix = '_combine_PTH_clusters'
```

```python
de_gene_set_vis = []
names_for_dotplot = []
num_vis = 8
for cluster in range(1,sdata.obs[clusters].unique().shape[0]+1):
    df = pd.read_csv(f'revision_results/primate_all_cells/pseudobulk_cluster_{cluster}_vs_rest{suffix}.csv', index_col=0)
    degs = df.sort_values('log2FoldChange', ascending=False).iloc[:20].index.tolist()
    if len(set(df.loc[degs[:num_vis]]['gene_name'].tolist())) < 8:
        de_gene_set_vis.extend(df.loc[degs[:num_vis+1]]['gene_name'].drop_duplicates().index.tolist())
        names_for_dotplot.extend(df.loc[degs[:num_vis+1]]['gene_name'].drop_duplicates().tolist())
    else:
        de_gene_set_vis.extend(degs[:num_vis])
        names_for_dotplot.extend(df.loc[degs[:num_vis]]['gene_name'].tolist())
```

```python
sns.set(font_scale=1.4, style='white')
```

```python
fig, ax =plt.subplots(figsize=(5,10))
axes = sc.pl.dotplot(sdata, groupby=clusters, return_fig=False, ax=ax, swap_axes=True, expression_cutoff=1e-3,
                     var_names=de_gene_set_vis, use_raw=False, standard_scale='var', show=False, cmap='Blues')
axes['mainplot_ax'].set_xticklabels(range(1,sdata.obs[clusters].unique().shape[0]+1), rotation=0)
axes['mainplot_ax'].set_yticklabels(names_for_dotplot)
    
plt.tight_layout()

fig.savefig(f'revision_results/primate_all_cells/markers{suffix}.png', dpi=500)
```
