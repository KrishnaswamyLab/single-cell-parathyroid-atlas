```python
import scanpy as sc
from sklearn.preprocessing import scale
import numpy as np
import graphtools, magic, scprep
```

```python
sc.set_figure_params(dpi_save=500)
```

## Primate pre-BC

```python
sdata = sc.read_h5ad('intermediate_files/3_primate_all_cells_clustering.h5ad')
```

```python
cluster_conv = {'0': '2', '1': '1', '2':'3', '3':'4', '4':'5'}
sdata.obs['clusters_res_0.1'] = [cluster_conv[x] for x in sdata.obs['clusters_res_0.1']]
```

```python
scprep.plot.scatter2d(sdata.obsm['X_phate_mnn_bmc'], c=sdata.obs['clusters_res_0.1'])
```

```python
cluster_conv = {'1': '1', '2': '1', '3':'2', '4':'3', '5':'4'}
sdata.obs[f'clusters_combined'] = [cluster_conv[x] for x in sdata.obs['clusters_res_0.1']]
```

```python
scprep.plot.scatter2d(sdata.obsm['X_phate_mnn_bmc'], c=sdata.obs['clusters_combined'])
```

```python
pbs = []
names = []
for cluster in [1,2,3,4]:
    for sample in sdata.obs['sample'].unique():
        sdata_subset = sdata[sdata.obs['sample'] == sample]
        samp_cell_subset = sdata_subset[sdata_subset.obs['clusters_combined'] == str(cluster)]

        rep_adata = sc.AnnData(X = samp_cell_subset.X.mean(axis=0), var=samp_cell_subset.var[[]])
        rep_adata.obs['sample'] = sample
        rep_adata.obs['cluster'] = int(cluster)
        names.append(f'{sample}_cluster{cluster}')
        pbs.append(rep_adata)
```

```python
pb = sc.concat(pbs)
pb.obs_names = names
```

```python
pb.X = scale(pb.X)
```

```python
g= sc.pl.clustermap(pb, row_cluster=False, col_cluster=True, figsize=(10,10), show=False, cmap='RdBu_r', xticklabels=False)
ax = g.ax_heatmap
ax.set_xlabel("All measured genes")
ax.set_ylabel("Samples")
g.savefig('revision_results/primate_all_cells/clustermap_primate_genome_scale_heatmap.png')
```

## Primate BC

```python
G = graphtools.Graph(sdata.obsm["X_pca_bmc"], use_pygsp=True, n_pca=None, 
                     random_state=42, sample_idx=sdata.obs.batch, bandwidth_scale=0.5)
```

```python
magic_op = magic.MAGIC(random_state=42, t=1)
magic_op.graph = G
sdata.layers['X_magic'] = magic_op.transform(sdata.to_df().values, ) 
```

```python
pbs = []
names = []
for cluster in [1,2,3,4]:
    for sample in sdata.obs['sample'].unique():
        sdata_subset = sdata[sdata.obs['sample'] == sample]
        samp_cell_subset = sdata_subset[sdata_subset.obs['clusters_combined'] == str(cluster)]
        samp_cell_subset.X = samp_cell_subset.layers['X_magic']

        rep_adata = sc.AnnData(X = samp_cell_subset.X.mean(axis=0), var=samp_cell_subset.var[[]])
        rep_adata.obs['sample'] = sample
        rep_adata.obs['cluster'] = int(cluster)
        names.append(f'{sample}_cluster{cluster}')
        pbs.append(rep_adata)
```

```python
pb = sc.concat(pbs)
pb.obs_names = names
```

```python
pb.X = scale(pb.X)
```

```python
g= sc.pl.clustermap(pb, row_cluster=False, col_cluster=True, figsize=(10,10), cmap='RdBu_r', xticklabels=False, show=False)
ax = g.ax_heatmap
ax.set_xlabel("All measured genes")
ax.set_ylabel("Samples")
g.savefig('revision_results/primate_all_cells/clustermap_primate_bc_genome_scale_heatmap.png')
```
